## 10. Find Total Number OF Job Title contain Captain 

```python
data['JobTitle'].str.contains('Captain',case=False).sum()
```

## 11. Display all The Names From Fire Department 

```python
data[data['JobTitle'].str.contains('Fair',case=False)][['EmployeeName','JobTitle','BasePay','OvertimePay','TotalPay']]
```

## 12. Replace 'Not Provided' in EmployeeName Column with Most common value

```python
data['BasePay'] = data['BasePay'].replace("Not Provided",data['BasePay'].mode()[0])
data['BasePay']
```

## 13. Find Minimum , Maximum and Average Basepay 

```python
data['BasePay'] = data['BasePay'].astype(float)
data['BasePay'].describe()
```

## 14. Drop The Row Having 5 Missing Values

```python
data = data[data.isnull().sum(axis=1) != 5] 
```

```python
data.isnull().sum(axis=1) 
```

## 15. Find Job Title of ALBERT PARDINI 

```python

data[data['EmployeeName'] =='ALBERT PARDINI']['JobTitle']
```

## 16. How Much ALBERT PARDINI Make (Including Benefits) ?

```python
data[data['EmployeeName'] == 'ALBERT PARDINI']['TotalPayBenefits']
```

## 17. Display Name of The Person Having Highest Basepay

```python
data[data['BasePay'].max() == data['BasePay']][['EmployeeName','BasePay']]
```

## 18. Find Average Basepay of All Employee per Year

```python
data['Year'] = pd.to_numeric(data['Year'],errors= 'coerce') 
```

```python
data.groupby('Year')['BasePay'].mean() 
```

## 19. Find Average BasePay of All Employee Per Job Title 

```python
data.groupby('JobTitle')['BasePay'].mean()
```

## 20. Find Average BasePay of All Employee Having Job Title ACCOUNTANT

```python
data[data['JobTitle'] == 'ACCOUNTANT']['BasePay'].mean()
```

## Find Average BasePay of All Employee Having Job Title All  ACCOUNTANT

```python
data[data['JobTitle'].str.contains('ACCOUNTANT')]['BasePay'].mean()
```
